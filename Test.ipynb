```python
import keras
from keras.models import load_model
import numpy as np
from keras.utils import plot_model
import scipy.io as sio
```

```python
x_data = np.load('test_longi_X_v1.0.npy')
t_data = np.load('test_longi_Y_v1.0.npy')

# x_data = np.load('test_trans_X_v1.0.npy')
# t_data = np.load('test_trans_Y_v1.0.npy')

print('size of test data is', np.shape(x_data)[0])
```

```python
model = load_model('D://Models2/longi_v1.0.h5')
model.summary()

plot_model(model, show_shapes = True, to_file='modelGraph.png')
```

```python
output = model.predict(x_data[:, :, :, :])
sio.savemat('output.mat', mdict={'input': x_data[:, :, :, 0].T, 'output': output[:, :, :, 0].T, 'ref': t_data[:, :, :, 0].T})
```
